## 教学(Teaching)

**教学方式(Teaching Approach)**

我将采用**自顶向下(top-down)**的教学法, 这与多数数学课的常见做法不同. 典型的**自底向上(bottom-up)**做法是: 先分别学会日后会用到的每个零件, 再逐步拼成更复杂的结构. 其问题在于学生容易失去动力, 缺少「大局观(big picture)」, 也不清楚自己究竟需要哪些知识. 

若你上过 fast.ai 的深度学习课, 用的就是这一套. 关于我的教学理念, 可见 [这篇博文](http://www.fast.ai/2016/10/08/teaching-philosophy/) 或 [这场演讲](https://vimeo.com/214233053). 

哈佛大学教授 David Perkins 著有 [*Making Learning Whole*](https://www.amazon.com/Making-Learning-Whole-Principles-Transform/dp/0470633719), 书中以棒球作类比: 我们不会要求孩子先背熟全部规则, 搞懂所有技术细节才允许上场; 相反, 他们先带着整体感觉开始打, 再随时间逐步补全规则与细节. 

总之, 一开始并非样样都懂很正常--本就不该一开始就全懂. 我们会先使用一些尚未解释的「黑箱(black boxes)」或**矩阵分解(matrix decompositions)**, 之后再深入底层细节. 

起步阶段请关注事物**做什么(DO)**, 而不是**它本质上是什么(ARE)**. 

人们通过以下方式学习: 
1. **动手做(doing)**(编程(coding)与搭建)
2. **讲解(explaining)**所学(写作或帮助他人)
